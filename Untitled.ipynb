```python
import pandas as pd
import glob
```

```python

path = r'C:\Users\beach\Desktop\tableau 2' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame
```

```python
dfclean=frame.dropna(axis=0)
```

```python
dfclean.dtypes
```

```python

frame1 = dfclean.gender.replace(to_replace=1, value='Male', regex=True)
frame2 = frame1.replace(to_replace=2, value='Female', regex=True)
frame3 = frame2.replace(to_replace=0, value="Unknown", regex=True)
```

```python

dfdrop = dfclean.drop(['gender'], axis=1)
```
